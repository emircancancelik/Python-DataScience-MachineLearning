```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2, "b")
my_axes.plot(data2, data1)
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1, data1 + 2, color="blue", linewidth=1.0)
new_axes.plot(data1, data1 + 3, color="red", linewidth=2.0)
new_axes.plot(data1, data1 + 6, color="blue", linestyle="--", marker="o", markersize=8, markerfacecolor="red")
new_axes.plot(data1, data1 + 8, color="blue", linestyle="--", marker="+", markersize=4)

plt.show()
```
