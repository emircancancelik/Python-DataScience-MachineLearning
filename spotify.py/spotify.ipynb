```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd

# 1. KİMLİK BİLGİLERİ
client_id = '568c2b777f644efdb8635856b9ebbd93'
client_secret = '46c38d7c909a48d2b4c4154bdb882075'
# PORTU 9000 YAPTIK (Dashboard'da da değiştir!)
redirect_uri = 'http://127.0.0.1:9000/callback' 
# SADECE ID KISMINI YAZDIK
playlist_id = '37i9dQZF1EJe6Wv35251' 

# 2. SPOTIFY BAĞLANTISI (open_browser=False manuel giriş sağlar)
auth_manager = SpotifyOAuth(client_id=client_id,
                            client_secret=client_secret,
                            redirect_uri=redirect_uri,
                            scope="playlist-read-private",
                            open_browser=False)

sp = spotipy.Spotify(auth_manager=auth_manager)

# 3. VERİ ÇEKME
results = sp.playlist_items(playlist_id)
tracks = results['items']

data = []
for item in tracks:
    track = item['track']
    if track:
        data.append({
            'id': track['id'],
            'name': track['name'],
            'artist': track['artists'][0]['name'],
            'added_by': item['added_by']['id']
        })

df = pd.DataFrame(data)

# 4. TEKNİK ÖZELLİKLERİ EKLEME
track_ids = df['id'].tolist()
audio_features = sp.audio_features(track_ids)
features_df = pd.DataFrame(audio_features)

final_df = pd.merge(df, features_df[['id', 'danceability', 'energy', 'valence', 'tempo']], on='id')

print("Veriler başarıyla çekildi!")
print(final_df.head())
```
