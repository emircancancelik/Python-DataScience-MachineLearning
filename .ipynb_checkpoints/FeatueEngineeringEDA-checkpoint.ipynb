```python
# Sadece metin içeren sütunları otomatik seç
categorical_features = df_clean.select_dtypes(include=['object']).columns.tolist()

# Sadece sayı içeren sütunları otomatik seç
numeric_features = df_clean.select_dtypes(include=['number']).columns.tolist()

print("Kategorik listem:", categorical_features)
print("Sayısal listem:", numeric_features)
```

```python
print(numeric_features)
```

```python
print(categorical_features)
```

```python
# Sayı olması gereken sütunları belirle
true_numerical = ['Rating', 'Reviews', 'Size', 'Installs', 'Price', 'Day', 'Month', 'Year']

for col in true_numerical:
    # Sayıya çevrilemeyen karakterleri (M, k, +, ,) temizlemek gerekebilir
    # Şimdilik en temel haliyle sayıya zorlayalım:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

# Şimdi Pandas'a otomatik ayırttıralım
numeric_features = df_clean.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = df_clean.select_dtypes(include=['object', 'category']).columns.tolist()

print(f"Gerçek Sayısal Sütunlar: {numeric_features}")
print(f"Gerçek Kategorik Sütunlar: {categorical_features}")
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Numerical Features', fontsize=18, fontweight='bold')

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df_clean[numeric_features[i]],fill=True, color='b')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
plt.show()
```
