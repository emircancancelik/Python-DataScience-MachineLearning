```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("6-bank_customers.csv")
```

```python
df.columns
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
X = df.drop("subscribed", axis= 1)
y = df["subscribed"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
y_pred = logistic.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print("score", score)
print(classification_report(y_pred, y_test))
print("confusion matrix: \n", confusion_matrix(y_pred, y_test))     
```

# hyperparameter tuning

```python
model = LogisticRegression()
```

```python
penalty = ["l1","l2","elasticnet"]
c_values = [100, 10, 1, 0.1, 0.01]
solver = ["newton-cg", "lbfsg", "liblinear", "sag", "saga", "newton-cholesky"]
```

```python
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
params
```

```python
#gird search cv
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
gird = GridSearchCV(estimator = model, param_grid= params, cv=cv, scoring = "accuracy", n_jobs=-1)
```

```python
gird
```

```python
gird.fit(X_train, y_train)
```

```python
gird.best_params_
```

```python
gird.best_score_
```

```python
y_pred = gird.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
print("score", score)
print(classification_report(y_pred, y_test))
print("confusion matrix: \n", confusion_matrix(y_pred, y_test))     
```
