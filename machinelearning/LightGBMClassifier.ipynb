```python
# feature engineering
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(["deck", "embark_town", "alive"], axis=1)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["age"] = df["age"].fillna(df["age"].median())
```

```python
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])
```

```python
df["embarked"].value_counts()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["adult_male"] = df["adult_male"].astype(int)
df["alone"] = df["alone"].astype(int)
```

```python
df["alone"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop("survived", axis=1)
y=df["survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_cols = ["sex", "class", "embarked", "who"]
preprocessor = ColumnTransformer(transformers= [
    ("cat", OneHotEncoder(drop="first", handle_unknown = "ignore"), categorical_cols)
], remainder="passthrough")
X_train=preprocessor.fit_transform(X_train)
X_test=preprocessor.transform(X_test)
```

```python
encoded_cols = preprocessor.get_feature_names_out()
```

```python
X_train = pd.DataFrame(X_train, columns=encoded_cols)
X_test = pd.DataFrame(X_test, columns = encoded_cols)
```

```python
X_train
```

```python
%pip install lightgbm
```

```python
import lightgbm as lgb
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
clf = lgb.LGBMClassifier(verbosity = -1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print(classification_report(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
```

```python
importances = clf.feature_importances_

feature_names = X_train.columns
feature_importance = pd.DataFrame({
    "Feature": feature_names,
    "Importance": importances


    
}
                                 ).sort_values(by="Importance", ascending=True)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
lgb_model = lgb.LGBMClassifier(verbosity=-1)
```

```python
param_grid = {
    "n_estimators": [100, 300, 500, 1000],
    "max_depth": [3, 5, 7, None], # None yerine -1 de yazılabilir
    "learning_rate": [0.01, 0.05, 0.1, 0.3],
    "num_leaves": [15, 31, 63, 127],
    "min_child_samples": [10,20,30],
    "subsample": [0.6, 0.8, 1.0],
    "colsample_bytree": [0.6, 0.8, 1.0]
}
```

```python
random_search = RandomizedSearchCV(estimator = lgb_model, param_distributions=param_grid, cv=5, scoring="accuracy",
                                   verbose= 1, n_jobs=-1)
```

```python
random_search.fit(X_train, y_train)
```

```python
random_search.best_params_
```

```python
y_pred = random_search.predict(X_test)
```

```python
print(classification_report(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
```
