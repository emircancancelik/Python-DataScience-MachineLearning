```python
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, MinMaxScaler, RobustScaler
from sklearn.model_selection import RandomizedSearchCV
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
from sklearn.linear_model import Lasso, Ridge, ElasticNet, RidgeCV, LassoCV, ElasticNetCV
import plotly.express as px
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.svm import SVC, SVR
from sklearn.compose import ColumnTransformer
from lazypredict.Supervised import LazyRegressor
from sklearn import datasets
from sklearn.utils import shuffle
from sklearn.pipeline import Pipeline
```

```python
df["Target_Price"] = df["Close"].shift(-1) # tomorrow regression price.
# write will in target price column.
```

```python
df.tail()
```

```python
df['Target_Direction'] = (df['Target_Price'] > df['Close']).astype(int)
```

```python
df.dropna(inplace=True)
```

```python
X = df.drop(['Date', 'Target_Price', 'Target_Direction', 'Close'], axis=1)
y_reg = df["Target_Price"]
y_class = df["Target_Direction"]
X_train, X_test, y_train_reg, y_test_reg = train_test_split(X, y_reg, 
                        test_size=0.2, random_state=42, shuffle=False)
X_train, X_test, y_train_class, y_test_class = train_test_split(X, y_class, test_size=0.2, 
                                        random_state=42, shuffle=False)
```

```python
reg_tree = DecisionTreeRegressor(max_depth=5, random_state=42)
reg_tree.fit(X_train, y_train_reg)
reg_preds = reg_tree.predict(X_test)
```

```python
print("R2 Skoru: ", r2_score(y_test_reg, reg_preds))
print("MAE (Ortalama Mutlak Hata): ", mean_absolute_error(y_test_reg, reg_preds))
```

```python
class_tree = DecisionTreeClassifier(max_depth=5, random_state=42)
class_tree.fit(X_train, y_train_class)
class_preds = class_tree.predict(X_test)
```

```python
print("Doğruluk Oranı (Accuracy): ", accuracy_score(y_test_class, class_preds))
print("\nSınıflandırma Raporu:\n", classification_report(y_test_class, class_preds))
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
reg_models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "ElasticNet": ElasticNet(),
    "KNN Regressor": KNeighborsRegressor(n_neighbors=5),
    "SVR (Destek Vektörleri)": SVR(kernel='rbf'),
    "Decision Tree Reg": DecisionTreeRegressor(max_depth=5)
}
```

```python
for name, model in reg_models.items():
    model.fit(X_train_scaled, y_train_reg)
    preds = model.predict(X_test_scaled)
    r2 = r2_score(y_test_reg, preds)
    mae = mean_absolute_error(y_test_reg, preds)
    print(f"{name:25} | R2: {r2:8.4f} | MAE: {mae:6.2f}")
```

```python
class_models = {
    "Logistic Regression": LogisticRegression(),
    "Gaussian Naive Bayes": GaussianNB(),
    "KNN Classifier": KNeighborsClassifier(n_neighbors=5),
    "SVC (Destek Vektörleri)": SVC(probability=True),
    "Decision Tree Class": DecisionTreeClassifier(max_depth=5)
}
```

```python
for name, model in class_models.items():
    model.fit(X_train_scaled, y_train_class)
    preds = model.predict(X_test_scaled)
    acc = accuracy_score(y_test_class, preds)
    
    if hasattr(model, "predict_proba"):
        prob_preds = model.predict_proba(X_test_scaled)[:, 1]
        auc = roc_auc_score(y_test_class, prob_preds)
    else:
        auc = 0.0
        
    print(f"{name:25} | Accuracy: {acc:6.4f} | ROC-AUC: {auc:6.4f}")
```
