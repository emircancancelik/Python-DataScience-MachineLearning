```python
def get_age(name):
    match = re.search(r'(\d+)\s*(year|yo|years|y\.o)', str(name), re.IGNORECASE)
    
    if match:
        return int(match.group(1))
    else:
        return None
```

```python
df = df.dropna(subset = ['price'])
```

```python
df = df.copy()
df['age'] = df['name'].apply(get_age)
```
