```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer(as_frame=True)
```

```python
df = data.frame
```

```python
df.tail()
```

```python
df.shape
```

```python
data.feature_names
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
columns = data.feature_names
```

```python
X_train = pd.DataFrame(X_train, columns = columns)
```

```python
X_train
```

```python
X_test = pd.DataFrame(X_test, columns = columns)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
```

```python
logistic = LogisticRegression()
gbc = GradientBoostingClassifier()
```

```python
logistic.fit(X_train, y_train)
gbc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("log")

y_pred_log = logistic.predict(X_test)
print(accuracy_score(y_test, y_pred_log))
print(confusion_matrix(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))

print("------------")
y_pred_gbc = gbc.predict(X_test)
print(accuracy_score(y_test, y_pred_gbc))
print(confusion_matrix(y_test, y_pred_gbc))
print(classification_report(y_test, y_pred_gbc))
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
X_train_pca = pd.DataFrame(X_train_pca, columns = ["PCA 1", "PCA 2"])
```

```python
X_test_pca = pd.DataFrame(X_test_pca, columns = ["PCA 1", "PCA 2"])
```

```python
print("log")

y_pred_log = logistic.predict(X_test_pca)
print(accuracy_score(y_test, y_pred_log))
print(confusion_matrix(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))

print("------------")
y_pred_gbc = gbc.predict(X_test_pca)
print(accuracy_score(y_test, y_pred_gbc))
print(confusion_matrix(y_test, y_pred_gbc))
print(classification_report(y_test, y_pred_gbc))
```
