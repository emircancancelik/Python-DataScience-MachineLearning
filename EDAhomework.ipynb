```python
# tat ve karakter profilini belirleyenler:

# fixed acidity: üzümün kendisinden gelen tartarik ve malik asitlerdir. şaraba ferahlatıcı tadı verir.
# volatile acidity: şarabın asetik asit miktarı. fazla olmamalı şarabın aseton gibi kokmasına bozulmasına sebep olur
# citric acid: çok küçük miktarda bulunur ama şaraba tazelik ve meyvemsi tat katar.
# residual sugar: fermantasyon sonucu şarapta kalan şeker miktarıdır. şarabın sek(şekersiz) tatlı mı olacağını belirler.
```

```python
# koruyucular ve dayanıklılık:

# chlorides: şaraptaki tuz miktarı.
# free so2: şarabın içine eklenen bakteri üremesini engelleyen gaz.
# sulphates: so2 gazına dönüşerek antimikrobiyal koruma sağlayan 
# total so2: toplam kükürt gazı. fazlası yumurta gibi bir tat bırakır.
```

```python
# fiziksel ve yapısal özellikler:
# density: şarabın suya kıyasla yoğunluğu. şeker miktarı yoğunluk ile doğru orantılı. alkol oranı yoğunluk ile ters orantılı.
# pH: şarabın asitlik derecesi (3 ile 4 arasındadır) şarap ph düştükçe daha tart(ekşi) tat oluşur.
# alcohol: şarabın alkol yüzdesi. şarabın gövesini ve ağızda bıraktığı o sıcaklık ağırlık hissini belirler.
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,7))
correlation_matrix = winedata.corr()
sns.heatmap(correlation_matrix, annot = True, cmap='coolwarm', fmt='.2f')
plt.title("Wine Features With Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(12,7))

numeric_winedata = winedata.drop(columns = ['Id'], errors = 'ignore')
correlation_matrix = numeric_winedata.corr()

sns.heatmap(correlation_matrix, annot = True, cmap= 'RdYlGn', fmt='.2f', linewidths = 0.5)
plt.title("Wine Features Relation Maps")
plt.show()

plt.figure(figsize=(10,6))
sns.regplot(x='alcohol', y='quality', data=winedata, x_jitter=0.2, scatter_kws={'alpha':0.2}, line_kws={'color':'red'})
plt.title("Wine With Alcohol Linear Relation Maps")
plt.xlabel("Alcohol %")
plt.ylabel("Quality Point")                                                                               
plt.show()                                                                                
                                                                               
            
```
